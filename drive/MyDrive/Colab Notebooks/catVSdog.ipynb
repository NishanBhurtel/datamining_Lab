```python
import os
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Activation, Dropout, Flatten, Dense
from keras.layers import Conv2D, MaxPooling2D
from keras import optimizers
```

```python
img_size=(256,256)
batch_size=32
```

```python
dogvscat="/content/drive/MyDrive/CNN dataset/dogvscat"
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
    dogvscat + "/train",
    labels="inferred",
    label_mode="categorical",
    batch_size=batch_size,
    image_size=img_size,
    shuffle=True
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    dogvscat + "/test",
    labels="inferred",

    label_mode="categorical",
    batch_size=batch_size,
    image_size=img_size,
    shuffle=True
)
```

```python
model = Sequential()
```

```python
model=Sequential()
model.add(Conv2D(16,(3,3),input_shape=(256,256,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Conv2D(64,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Conv2D(128,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Conv2D(256,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
model.add(Dense(256))
model.add(Activation('relu'))
```

```python

model.add(Dense(24))
model.add(Activation('relu'))
```

```python

model.add(Dense(24))
model.add(Activation('relu'))
```

```python

model.add(Dropout(0.6))
model.add(Dense(2)) # Changed from 16 to 2
model.add(Activation('sigmoid'))
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)
]
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=0.0001),
              metrics=['accuracy'],

              )
```

```python
model.summary()
```

```python

history = model.fit(train_ds, epochs=20, validation_data=val_ds,callbacks=callbacks)
```
